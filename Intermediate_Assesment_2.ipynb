```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/train_LZdllcl.csv')
```

```python
data.head(100)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
graph=data.select_dtypes(include='float')
graph.hist(figsize=[10,7])
plt.show()
```

```python
data.columns
```

```python
for i in ['no_of_trainings', 'age', 'previous_year_rating',
       'length_of_service', 'KPIs_met >80%', 'awards_won?',
       'avg_training_score', 'is_promoted']:
    data[i]=data[i].fillna(data[i].median())
```

```python
data.isna().sum()
```

```python
for i in ['education', 'gender',
       'recruitment_channel']:
      data[i]=data[i].fillna(data[i].mode())
```

```python
data['education']=data['education'].fillna(data['education'].mode()[0])
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.columns
```

```python
correlation_matrix=data[['no_of_trainings', 'age', 'previous_year_rating',
       'length_of_service', 'KPIs_met >80%', 'awards_won?',
       'avg_training_score', 'is_promoted']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='YlGnBu',linewidths=2.0)
```

```python
x=data.drop(['employee_id','region','age'],axis=1)
```

```python
y=data['is_promoted']
```

```python
x=pd.get_dummies(x,dtype='int64')
```

```python
x.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
print('accuracy is',accuracy_score(y_test,y_pred))
print('precision is',precision_score(y_test,y_pred))
print('recall is',recall_score(y_test,y_pred))
print('f1 score is',f1_score(y_test,y_pred))
```
