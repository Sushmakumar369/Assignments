```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/car_age_price.csv')
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
corr_matrix=data.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corr_matrix,annot=True,color='Yellow')
```

```python
x=data.drop('Price',axis=1)
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
print('R squared value is :\n',r2_score(y_test,pred))
```

```python
data['Price']
```

```python
def predict_price(year):
    year_input = np.array(year).reshape(-1, 1)  # Ensure correct shape
    price_linear = model.predict(year_input)[0]
    return price_linear
```

```python
predicted_price_linear = predict_price([2022])
print(f"Predicted price using Linear Regression: {predicted_price_linear:.2f}")
```
